```python
import requests
import mysql.connector
import pandas as pd
from time import sleep
```

```python
def GetUsersList(page):
    r = requests.get('https://api.github.com/search/users?page=' + str(page) + '&q=location:russia+type:user?')
    return r.json()['items']
```

```python
df = pd.DataFrame.from_dict(GetUsersList(1))
for i in range(2, 10):
    df = df.append(pd.DataFrame.from_dict(GetUsersList(i)))
```

```python
def GetJSON(link):
    r = requests.get(link)
    return r.json()
```

```python
def GetNormilezedTime(time):
    time = time[:7]
    time = list(map(int, time.split('-')))
    time[0] = 2016 - time[0]
    time[1] = 12 - time[1]
    return time[0] * 12 + time[1]
```

```python
def GetLanguageExp(link):
    json = GetJSON(link)
    A = []
    for i in json:
        try:
            A.append(GetNormilezedTime(i['created_at']))
        except:
            A.append(0)
    A.sort()
    return A[len(A) - 1]
```

```python
df2 = df['repos_url'].apply(lambda x: GetLanguageExp(x))
```
